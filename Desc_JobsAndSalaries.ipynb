- Função dos métodos head() e tail():

  - head(n): Retorna as primeiras n linhas de um objeto. É ideal para verificar se os dados foram carregados corretamente, nomes de colunas e os primeiros registros.

  - tail(n): Retorna as últimas n linhas de um objeto. É útil para verificar o final do conjunto de dados, o que é crucial após operações de ordenação (sorting) ou para conferir a integridade dos dados mais recentes.

- Linhas exibidas por padrão:

  Ambos os métodos exibem, por padrão, as 5 primeiras (head) ou últimas (tail) linhas.

- Como especificar a quantidade de observações:

  Você pode passar um número inteiro como argumento dentro dos parênteses do método: df.head(10) exibe as primeiras 10 linhas, e df.tail(3) exibe as últimas 3 linhas.

- O que o método sample() permite observar?

  O método sample() seleciona registros aleatórios do DataFrame. Enquanto head() e tail() mostram dados ordenados de cima ou de baixo, sample() ajuda a observar a diversidade dos dados e encontrar outliers (valores discrepantes) ou padrões que podem não aparecer no topo ou no final do arquivo.

- Diferença entre head() e sample():

  - head(): Foca nos primeiros registros (ordem estrutural).

  - sample(): Seleciona linhas aleatórias, oferecendo uma visão "neutra" e não enviesada do conjunto de dados, o que é melhor para evitar conclusões precipitadas baseadas apenas no início do arquivo.

####**Problema do tipo *object***


🐍 No Python, **object** é o tipo base do qual praticamente todos os outros tipos derivam. No contexto do **Pandas**, quando uma coluna possui o tipo **object**, isso significa que ela armazena objetos Python genéricos. Na prática, esse tipo é frequentemente usado para armazenar textos ou dados em que o **Pandas** não conseguiu identificar um tipo mais específico. O **Pandas** utiliza **object** ao importar dados porque ele tenta inferir automaticamente o tipo das colunas; quando há texto, mistura de tipos ou incerteza na inferência, ele usa **object** como um tipo genérico. No entanto, o tipo **object** não garante que a variável seja categórica. Ele apenas indica que os valores são tratados como objetos Python. Para representar explicitamente variáveis categóricas, o tipo adequado no **Pandas** é category.

```python
df.select_dtypes(include="object").columns
```

- **Discussão:**

      No Pandas, o argumento **include="object"** (usado geralmente em select_dtypes)
      filtra apenas as colunas cujo tipo de dado é **object**.
      Essas colunas normalmente armazenam dados textuais ou dados
      que o **Pandas** não conseguiu classificar

      O atributo .columns retorna os nomes das colunas

      Sem .columns, o resultado exibido seria

      valores, e não apenas os nomes das colunas.

      todas as colunas que o Pandas classificou como object.
      Que neste caso são 9.

      Nem sempre. Algumas colunas podem dever
      ser convertidas para tipos mais apropriados, como:

      - category para variáveis categóricas;
      - datetime para datas;
      - tipos numéricos (int ou float) caso os valores representem números.

      Manter o tipo correto melhora uso de memória,
      desempenho e qualidade da análise de dados

##**Informações faltantes e duplicadas**
---






⛔Valores **nulos** são valores ausentes ou não preenchidos na sua base de dados, esse valores podem aparecer como:
* NULL
* NaN
* campo vazio
* “-”, “Não informado”, etc.

Exemplo:
| Nome | Idade | Salario |
|---|---|---|
|Ana|25|3000|
|João|NULL|2000|

Aqui, a idade de João está ausente.

👥 Valores **duplicados** são registros idênticos ou muito semelhantes que aparecem mais de uma vez, representando a mesma informação repetidamente.

Exemplo:
| Nome | idade | email |
|---|---|---|
|Ana|25|ana@email.com|
|João|30|joao@email.com|
|pedro|24|joao@email.com|

Aqui o email de joão esta duplicado no perfil de Pedro.

🔁**Duplicação de Linhas** (Registros Repetidos)
Acontece quando a mesma entidade (pessoa, produto, venda) aparece em várias linhas da tabela. É o erro de cadastro mais comum.

    Exemplo: O cliente "Carlos" cadastrado duas vezes com IDs diferentes, mas o mesmo CPF.

🔄**Duplicação de Colunas**(Redundância de Atributos)

Ocorre quando a mesma informação ou tipo de dado é espalhado por várias colunas na mesma tabela, geralmente por um erro de design.


    Exemplo: Uma tabela de "Produtos" com as colunas Telefone_Fornecedor_1, Telefone_Fornecedor_2 e Telefone_Fornecedor_3.

❓Esses aspectos devem ser analisados na
**Análise Exploratória de Dados (EDA)** porque eles garantem a integridade estatística e a qualidade da base, e se os dados de entrada forem ruidosos, incompletos ou repetitivos, qualquer conclusão derivado deles será enviesado e pouco confiável.

Problemas gerados caso não sejam tratados:

- **Distorção de Métricas**: Valores nulos e duplicatas alteram drasticamente cálculos essenciais, como médias, somas e desvios padrão, levando a interpretações de negócio completamente erradas.
- **Instabilidade em Modelos**: Dados duplicados (linhas ou colunas) causam overfitting ou multicolinearidade, fazendo com que algoritmos "viciem" em certas informações e percam a capacidade de prever novos cenários.
- **Desperdício de Recursos**: Processar informações redundantes ou campos vazios consome tempo de computação, memória e armazenamento desnecessariamente, tornando o fluxo de dados lento e ineficiente.

##**Padronização e Tradução de Variáveis**
---

```python
nivel_de_experiencia={
"Entry-level":"Júnior",
"Mid-level":"Pleno",
"Senior":"Sênior",
"Executive":"Executivo"
}

# Tradução: tipo de emprego
map_emprego = {
    "Full-time": "Tempo integral",
    "Part-time": "Meio período",
    "Contract": "Contrato",
    "Freelance": "Freelancer"
}

# Tradução: ambiente de trabalho
map_work = {
    "Remote": "Remoto",
    "Hybrid": "Híbrido",
    "In-person": "Presencial"
}

# Tradução: tamanho da empresa
map_empresa = {
    "S": "Pequena",
    "M": "Média",
    "L": "Grande"
}

df_reduzido["experience_level"]=df_reduzido["experience_level"].map(
nivel_de_experiencia)

df_reduzido["employment_type"] = df_reduzido["employment_type"].map(map_emprego)

df_reduzido["work_setting"] = df_reduzido["work_setting"].map(map_work)

df_reduzido["company_size"] = df_reduzido["company_size"].map(map_empresa)
```

```python
df_reduzido[["experience_level", "employment_type", "work_setting", "company_size"]].value_counts()
```

- **Discussão:**

      Um dicionário em Python é uma estrutura que armazena pares chave →
      valor. Nesse código, ele é usado para associar os níveis de experiência
      em inglês aos seus equivalentes em português.

      O método map() aplica esse dicionário à coluna, substituindo cada valor
      original pelo correspondente definido no dicionário.

      Se algum valor da coluna não estiver presente no dicionário, ele será
      convertido para NaN, ou seja, ficará como valor ausente.

      A tradução é contextual, não apenas literal, pois adapta os termos para
      algo mais comum no contexto profissional brasileiro (ex: “Mid-level” →
      “Pleno”).

      Não é sempre recomendável substituir a coluna original. O ideal, em
      muitos casos, é criar uma nova coluna traduzida, para manter os dados
      originais e evitar perda de informação.

####**Ordenação**


```python
ordem=["Júnior", "Pleno", "Sênior", "Executivo"]

df_reduzido["experience_level"]=pd.Categorical(
df_reduzido["experience_level"],
categories=ordem,
ordered=True
)

df_reduzido.sort_values("experience_level").head()
```

- **Discussão:**

      A lista ordem define a hierarquia dos níveis de experiência, ou seja, a
      ordem correta do menor para o maior nível profissional.

      Definir ordered=True significa que essa variável categórica agora tem
      ordem lógica, permitindo comparações e ordenações baseadas nessa
      hierarquia.

      O pd.Categorical() transforma a coluna em uma variável categórica
      ordenada, usando a lista definida como referência de ordem.

      Essa operação não altera os valores em si, apenas muda a forma como o
      pandas interpreta e organiza essa variável.

      O método sort_values() ordena os dados com base na coluna informada.
      Nesse caso, ele usa a hierarquia definida, e não a ordem alfabética.

      A diferença é:

      Ordem alfabética: “Executivo” viria antes de “Júnior” (errado no
      contexto)
      Ordem hierárquica: segue a progressão real (Júnior → Pleno → Sênior →
      Executivo)

      Isso é importante em análises estatísticas porque garante resultados
      mais coerentes, principalmente em gráficos, comparações e agrupamentos
      por nível de experiência.

- O que é um dicionário em Python e por que ele é adequado para essa tarefa?

📕Um **dicionario** em phyton é uma estrutura de dados fundamental, não ordenanda, que armazena elementos em chave-valor("work_year" seria a chave, "ano_trabalho" seria o valor). Neste caso essa estrutura é adequada porque permite mapear nomes originais para tradução de forma direta, garante organização e padronização, facilita consultas e automatizações é simples de entender e manter.

- O que faz o método rename(columns=...) do pandas?

🎭O método **rename(columns=...)** é usado para renomear colunas de um **DataFrame** Ele permite substituir os nomes atuais das colunas por novos nomes, normalmente usando um **dicionário**, que foi exatamente o que fizemos no código. Exemplo:
**"nome_antigo" : "nome_novo"**

- O dataframe original é modificado ou um novo dataframe é criado? Explique

👜Neste caso um novo **DataFrame** é criado, já que no código referenciamos as tabelas que queremos modificar os nomes, porém na hora de fazer a alteração o **Pandas** normalmente não faz alteração direto no **DataFrame** para evitar alterações inesperadas nos dados. Então para que acontecesse a alteração direta no **DataFrame** teriamo que adicionar ao código **"inplace=True"(que altera no próprio data frame)**. E então o código ficaria **"df.rename(columns=novos_nomes, inplace=True)"**

- Por que é uma boa prática manter os nomes das colunas padronizados em um idioma único?

🧐Essa pratica é conciderada uma boa prática porque melhora a organização, clareza e manutenção do projeto, já que evita confusões e ambiguidade, por exemplo linhas com as mesmas informações só que em linguagens diferentes  

##**Estatísticas descritivas**
---


📈Para retirar uma **Estatistica Descritiva** do nosso Banco de Dados, podemos utilizar um método do **Pandas** denominado de **"describe()"**, que pode gerar um resumo estatístico rápido de DataFrames ou Séries. O cálculo depende do que o Pandas encontra na coluna:
- Para Números: Contagem, Média, Desvio Padrão, Mínimo, Máximo e os Quartis(25%, 50% e 75%).
- Para Objetos/Categorias: Contagem, Quantidade de valores únicos, o valor mais frequente (top) e sua frequência.

O método é versátil e se aplica a quase todos os tipos de dados comuns em Data Science:
- Numéricas: int64 (inteiros) e float64 (decimais).
- Categóricas/Texto: object ou category.
- Temporais: datetime64 (datas e horários).

      Observação: Por padrão, se o seu DataFrame tiver colunas
      numéricas e de texto, o describe()
      mostrará apenas as numéricas.
      Para ver as outras, você deve usar o comando:
       df.describe(include='all').


Para interpretar os resultados, é fundamental entender o que cada métrica diz sobre seus dados:

|Estatística|	Significado|
|---|---|
|Média|É o centro de massa dos dados. Calculada somando todos os valores e dividindo pela quantidade. É sensível a valores muito altos ou baixos (outliers).|
|Mediana (50%)|É o valor que divide os dados exatamente ao meio. Se você ordenar os dados, 50% estarão abaixo e 50% acima dele. Diferente da média, ela não é afetada por valores extremos.|
|Desvio Padrão|	Indica o grau de "dispersão". Um desvio baixo significa que os valores estão próximos da média; um desvio alto significa que os dados estão espalhados em uma faixa ampla.|
|Mínimo|O menor valor registrado na coluna. Ajuda a identificar o limite inferior ou possíveis erros (ex: idade negativa).|
|Máximo|O maior valor registrado na coluna. Ajuda a identificar o limite superior ou picos nos dados.|

##**Comparação de colunas de salário**
---

🪙A diferença entre o salário registrado na moeda original e o salário convertido para uma moeda padrão, como o dólar americano (USD), está principalmente na possibilidade de comparação. Quando o salário é apresentado na moeda do próprio país, ele representa com fidelidade a realidade econômica local. Porém, esse valor isolado não permite comparações diretas entre diferentes países, pois cada país possui sua própria moeda, com taxas de câmbio, níveis de inflação e poderes de compra distintos.

A padronização monetária é essencial para análises comparativas internacionais porque:
- Garante coerência estatística
- Permite decisões estratégicas fundamentadas
- Facilita interpretações claras e objetivas

Em Ciência de Dados, comparar exige primeiro padronizar.


```python
df_reduzido[['salario','salario_dolar','moeda_salario']].head(10)
```

```python
df_reduzido['moeda_salario'].value_counts()
```

- **Discussão**

      As colunas salario e salario_dolar não representam exatamente a mesma
      informação. A primeira está na moeda original de cada registro, enquanto
      a segunda já está convertida para dólar, permitindo comparação direta.

      A variável moeda_salario é essencial porque indica em qual moeda o
      salário original foi registrado, sendo necessária para interpretar
      corretamente os valores e entender as conversões.

      A quantidade de moedas diferentes pode ser vista com value_counts(). Esse
      resultado mostra a diversidade de moedas presentes no banco, indicando
      que os dados vêm de diferentes países.

      Converter salários para uma moeda padrão (USD) é importante porque

      moeda_salario.
      
      O resultado mostra quais moedas são mais frequentes e quantas existem no

##**Tipos de gráficos e sua relação com os dados**
---

####**Gráfico de barras**
📊O gráfico de barras é uma ferramenta visual que representa dados categóricos através de barras retangulares, onde o comprimento/altura é proporcional ao valor que representam. Essencial para comparar categorias, ele mapeia categorias no eixo x e valores numéricos no eixo y (ou vice-versa), facilitando a visualização rápida de diferenças, tendências e frequências.
Principais Aspectos da Relação Dados e Gráfico:

- Estrutura de Dados: Utiliza dados categóricos (nomes, produtos, meses) e dados quantitativos (vendas, quantidades, frequências).
- Representação: A altura (barras verticais) ou o comprimento (barras horizontais) denota a medida do dado.

Exemplo:

<img src="https://observatorioobstetricobr.org/wp-content/uploads/2021/12/fig2.png" width="500">

####**Gráfico de linhas**

📉 Um tipo de gráfico que exibe informações como uma série de pontos de dados chamados de 'marcadores', conectados por segmentos de linhas retas. É a ferramenta padrão para visualizar sequências ordenadas.

Relação com os dados:
- Continuidade: Ele estabelece uma conexão lógica entre os pontos, tratando os dados como uma progressão e não como categorias isoladas.
- Tendência: Transforma números brutos em movimento visual (ascensão, queda ou ciclos), facilitando a identificação de padrões ao longo do tempo.
- Densidade: Permite relacionar grandes volumes de dados (como cotações minuto a minuto) de forma compacta e fluida.

Exemplo:

<img src="https://www.jaspersoft.com/content/dam/jaspersoft/images/graphics/infographics/line-chart-example.svg" width="700">

####**Gráfico de Pizza**
🍕 É um gráfico circular dividido em fatias, onde o círculo inteiro representa o total (100%) e cada fatia representa uma parte desse todo. É uma ferramenta puramente estatística de composição.

Relação com os dados:
- Proporcionalidade: A relação principal é a da parte com o todo. Ele converte valores numéricos em áreas proporcionais (ângulos), permitindo que o usuário entenda instantaneamente qual categoria domina o conjunto.
- Composição Percentual: Ele traduz dados absolutos em percentuais. Diferente das barras, onde comparamos alturas, aqui comparamos a fatia em relação ao círculo completo.
- Baixa Granularidade: Relaciona-se melhor com poucos dados (idealmente até 5 ou 6 categorias). Quando há muitas divisões, a relação visual se perde e os dados tornam-se ilegíveis.

 Exemplo:

<img src="https://docs.oracle.com/applications/help/pt_BR/enterprise-performance-management/11.2/HFWCC/img/pie_chart_example_2.jpg" width="500">

####**Boxplot**
📦 É uma ferramenta estatística usada para representar a variação e a distribuição de um conjunto de dados numéricos por meio de quartis. Ele é composto por uma "caixa" central e linhas (chamadas de bigodes) que se estendem para fora dela.

Relação com os dados:
- Distribuição Estatística: Ele divide os dados em quatro partes iguais (25% cada). A relação visual não é sobre a soma dos valores, mas sobre onde a maioria dos dados está concentrada ou dispersa.
- Identificação de Outliers: Relaciona pontos individuais que estão muito fora da curva (valores atípicos), exibindo-os como pontos isolados além dos limites dos bigodes.
- Mediana vs. Média: A linha central dentro da caixa mostra a mediana, revelando se os dados estão "inclinados" para valores mais altos ou mais baixos (assimetria).
- Amplitude e Variabilidade: O tamanho da caixa (Intervalo Interquartil) mostra o quão consistentes ou voláteis são os dados; uma caixa curta indica dados muito parecidos, enquanto uma longa indica alta variação.

Exemplo:

<img src="https://fernandafperes.com.br/blog/interpretacao-boxplot/g2.png" width="500">



####**Gráfico de Dispersão**

<img src="https://cdn-icons-png.flaticon.com/128/7665/7665284.png" width="16">É um gráfico que utiliza coordenadas cartesianas para exibir valores de duas variáveis para um conjunto de dados. Cada ponto no gráfico representa uma observação individual, posicionada de acordo com seus valores nos eixos horizontal (X) e vertical (Y).

Relação com os dados:
- Correlação: Sua função principal é revelar a relação entre duas variáveis numéricas. Ele mostra se, quando uma variável aumenta, a outra também aumenta (correlação positiva), diminui (correlação negativa) ou se não há relação aparente.
- Distribuição e Agrupamento: Permite identificar visualmente "nuvens" de pontos (clusters), mostrando onde a maioria dos dados se concentra ou se existem lacunas entre eles.
- Independência: Ao contrário do gráfico de linhas, o de dispersão não conecta os pontos, pois a ordem entre eles não importa. A relação é puramente sobre a posição relativa de uma variável em função da outra.
- Causalidade e Previsão: É o ponto de partida para análises de regressão, onde se tenta traçar uma linha média que explique como os dados se relacionam matematicamente.
